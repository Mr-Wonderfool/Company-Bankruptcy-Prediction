充分性检测
检验总体变量的相关矩阵是否是单位阵（相关系数矩阵对角线的所有元素均为1,所有非对角线上的元素均为零），即检验各个变量是否各自独立。
如果不是单位矩阵，说明原变量之间存在相关性，可以进行因子分子；反之，原变量之间不存在相关性，数据不适合进行主成分分析

KMO检验
检查变量间的相关性和偏相关性，取值在0-1之间；KOM统计量越接近1，变量间的相关性越强，偏相关性越弱，因子分析的效果越好。
通常取值从0.6开始进行因子分析

```python
import pandas as pd
import numpy as np
from factor_analyzer import FactorAnalyzer
import factor_analyzer
from sympy import *
from matplotlib import pyplot as plt


#计算相关系数矩阵
data_corr=factor_analyzer.corr(df)
print(data_corr)

#验证是否适用于因子分析
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(df)
print(kmo_model)

from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value, p_value = calculate_bartlett_sphericity(df)
chi_square_value, p_value
```

选择因子个数,计算相关矩阵的特征值，进行降序排列

```python
faa = FactorAnalyzer(25,rotation=None)
faa.fit(df)
# 得到特征值ev、特征向量v
ev,v=faa.get_eigenvalues()
print(ev,v)
```

```python
# 同样的数据绘制散点图和折线图
plt.scatter(range(1, df.shape[1] + 1), ev)
plt.plot(range(1, df.shape[1] + 1), ev)
 
# 显示图的标题和xy轴的名字
plt.title("Scree Plot")  
plt.xlabel("Factors")
plt.ylabel("Eigenvalue")
 
plt.grid()  # 显示网格
plt.show()  # 显示图形
```

```python
# varimax 方差最大化
# 选择固定因子为 29 个
faa = FactorAnalyzer(29,rotation='varimax')
faa.fit(df)
```

转成新变量，将原始数据转成29个新的特征

```python
#faa.transform(df)
df2 = pd.DataFrame(faa.transform(df))
print(df2)
```

汇总得分
使用各个因子方差贡献率占29个因子方差贡献率的比重作为权重进行加权汇总，得到各公司的综合得分F

```python
from pandas import DataFrame,Series
def F(factors):
    return sum(factors*faa.get_factor_variance()[1])

scores = []
data = DataFrame(df)
for i in range(len(faa.transform(data))):
    new = F(faa.transform(data)[i])
    scores.append(new)

print(scores)
```

```python
data['综合打分'] = scores
data = data.sort_values(by='综合打分',ascending=False)
print(data['综合打分'])
data.to_csv('addscore.csv', index=False)
```
